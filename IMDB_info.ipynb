```python
# Import packages
from bs4 import BeautifulSoup
import urllib.request
import csv
import re
```

```python
start=1 #Start from the first page
rawlink=[]  # Create an empty list 
```

```python
#Each page includes 50 movies.To get 200 movies,we should stop at start=151
while (start<=200): 
    url ='https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=horror&sort=user_rating,desc&start='+str(start)+'&ref_=adv_nxt'
    ourUrl = urllib.request.urlopen(url)
    soup = BeautifulSoup(ourUrl,'html.parser')

    for i in soup.find_all('h3',{'class':'lister-item-header'}):  
        per_link=i.find('a')  
        rawlink.append(per_link)  
    start=start+50
```

```python
Movie = [] #Build the movie list
for i in rawlink:
    i = str(i).replace('</a>','')
    per_movie = i[28:]
    Movie.append(per_movie)
print(Movie)
```

```python
link=[] #Build the linklist for each movie
for i in rawlink:
    strlist=re.compile(r'\"(.*)\"')
    name=strlist.findall(str(i))[0]
    print(name)
    fullname="https://www.imdb.com"+name
    link.append(fullname)
```

```python
f = open('Web_IMDB_comments.csv','w',encoding='utf-8',newline='')
csv_writer = csv.writer(f)
csv_writer.writerow(["Name","Rating","Title","Comment"])
i=1 #i represents the number for the movie.
while i <= 199:
    webrating=1 #Webrating represents the rating of each comments. From 1 to 10.
    while (webrating<=10):
        url =link[i]+"reviews?sort=helpfulnessScore&dir=desc&ratingFilter="+str(webrating)
        ourUrl = urllib.request.urlopen(url)
        soup = BeautifulSoup(ourUrl,'html.parser')
        #Reviews are divided into 3 parts: ratings, titles and contents
        #Rating is termined by webrating.
        #Now build the Title list for each page.
        Title = []
        for s in soup.find_all('a',{'class':'title'}):
            sentences = str(s).rstrip('\n</a>')
            sentences=re.sub("(.*>)","",sentences,1)
            Title.append(sentences)
        #Now build the comments list for each page.
        comments = []
        for s in soup.find_all('div', {'class': 'content'}):
            comment = s.find('div')
            comment = str(comment).rstrip('</div>')
            comment=re.sub("(.*?>)","",comment,1)
            comment=comment.replace("<br/>","")
            comment=comment.replace("\n"," ")
            comments.append(comment)
        #Transfer the comment information into csv file.
        for j in range(len(Title)):
            csv_writer.writerow([Movie[i],webrating,Title[j],comments[j]])
        print(Movie[i],i,webrating)
        webrating=webrating+1
    i=i+1
        
f.close()
```
