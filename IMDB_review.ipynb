```python
# Import packages
from bs4 import BeautifulSoup
import urllib.request
import csv
import re
```

```python
start=1 #Start from the first page
rawlink=[]  # Create an empty list for rawlink
```

```python
#Each page includes 50 movies.To get 200 movies,we should stop at start=151
while (start<=200): 
    url ='https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=horror&sort=user_rating,desc&start='+str(start)+'&ref_=adv_nxt'
    ourUrl = urllib.request.urlopen(url)
    soup = BeautifulSoup(ourUrl,'html.parser')

    for i in soup.find_all('h3',{'class':'lister-item-header'}):  
        per_link=i.find('a')  
        rawlink.append(per_link)  
    start=start+50
```

```python
#Build the linklist for each movie
link=[]
for i in rawlink:
    strlist=re.compile(r'\"(.*)\"')
    name=strlist.findall(str(i))[0]
    print(name)
    fullname="https://www.imdb.com"+name
    link.append(fullname)
```

```python
#extract moviename
for i in rawlink:
    i = str(i).replace('</a>','')
    per_movie = i[28:]
    print(per_movie)
```

```python
f = open('Web_Project_imdbmovies.csv','w',encoding='utf-8',newline='')
csv_writer = csv.writer(f)
csv_writer.writerow(['Name','Rating','Year','Duration','Director','Writer','Actor'])
```

```python
for num in range(len(rawlink)):
    cast = []
    url=link[num]
    ourUrl = urllib.request.urlopen(url)
    soup = BeautifulSoup(ourUrl,'html.parser')
    #extract name
    name=str(rawlink[num]).replace('</a>','')
    name=name[28:]
    # extract rating
    for i in soup.find_all('div',{'class': 'ratingValue'}):
        rating = i.find('span').text
    # extract year
    for i in soup.find_all('span',{'id': 'titleYear'}):
        year = i.find('a').text
    # extract duration
    for i in soup.find_all('div',{'class':'subtext'}):
        b = i.find('time').text
        duration = str(b).strip()
    
    # extract cast
    icast = []
    for i in soup.find_all('div',{'class': 'credit_summary_item'}):
        b = i.find_all('a')
        icast.append(b)
        cast.append(icast)
    #extract director from cast list
    director = []
    for i in cast[0][0]:
        i = str(i)
        per_director=i[27:].rstrip('</a>]')
        director.append(per_director)
    if "credit" in director[-1]:
        del director[-1]
    #extract writer rom cast list
    writer = []
    for i in cast[0][1]:
        i = str(i)
        i=i.replace("</a>","")
        per_writer=i[27:]
        writer.append(per_writer)
    if "credit" in writer[-1]:
        del writer[-1]
    #extract actor rom cast list
    actor = []
    for i in cast[0][2]:
        i = str(i)
        i=i.replace("</a>","")
        per_actor = i[27:]
        actor.append(per_actor)
    del actor[-1]
    print(num)
    csv_writer.writerow([name,rating,year,duration,director,writer,actor])
f.close()
```
